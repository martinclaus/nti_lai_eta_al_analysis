```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

from utils import (
    detrend, standardize, monthly_anomalies
)
```

```python
eq = dict(lat=0,lon=slice(140,260))
```

#### 1.1 NEMO SSH 

```python
t_grid_file = xr.open_zarr('data/ORCA025.L46-KFS006_TroPA.zarr')
ssh_fitting = t_grid_file.sossheig.sel(**eq)
```

```python
ssh_nemo_eq = detrend(monthly_anomalies(ssh_fitting).compute())
plt.figure(figsize=(6, 8))
ssh_nemo_eq.T.plot()
```

#### 1.2 linear SSH 

```python
eta = xr.open_zarr('data/lmmm_eta_nemo.zarr').eta
eta_fitting = (
    eta
    .assign_coords(
        {'time': (eta.time-np.timedelta64(1314900,'s')).data}
    )
    .sel(lon=eq["lon"])
    .interp(lat=0., time=ssh_nemo_eq.time)
    .fillna(0.)
    .compute()
)
```

```python
plt.figure(figsize=(6, 8))
eta_fitting.sel(mode=2).plot();
```

```python
# fit to full data
def extract_data(t_slice, max_mode=10):
    X = (
        eta_fitting
        .isel(time=t_slice, mode=slice(max_mode))
        .stack(sample=("time", "lon"))
        .T
    )
    y = (
        ssh_nemo_eq
        .isel(time=t_slice)
        .stack(sample=("time", "lon"))
    )
    return X, y
```

```python
coef = np.zeros((len(eta_fitting.mode), len(eta_fitting.mode)))
for i, max_mode in enumerate(range(1, 11)):
    X, y = extract_data(slice(743), max_mode)
    reg = LinearRegression(fit_intercept=False).fit(X, y)
    coef[i, :max_mode] = reg.coef_    
with np.printoptions(precision=3, suppress=True):
    print(coef)
```

```python
mod1 = coef[2,0]*eta_fitting.sel(mode=1)
mod2 = coef[2,1]*eta_fitting.sel(mode=2)
mod3 = coef[2,2]*eta_fitting.sel(mode=3)
m3=(mod1+mod2+mod3)*-1
```

```python
ssh_lin_eq = detrend(m3)
```

#### 1.3 D20 data 

```python
d20_mon_anom = monthly_anomalies(
    (-1) * t_grid_file.d20.sel(**eq)
)
d20_mon_anom.attrs.update({"long_name": "D20 anomaly", "units": "m"})

d20_eq = detrend(
    d20_mon_anom.fillna(0.).compute()
).rename('D20')
```

```python
d20_eq.plot()
```

#### 1.4 SST 

```python
sst_o = t_grid_file.sosstsst.sel(**eq)
```

```python
sst_mon = (
    sst_o
    .resample(time="1M",label="left",loffset="15d")
    .mean("time")[:-1]
).squeeze().compute()
sst_mon.attrs["units"] = "C"
```

```python
sst_mon_grouped = sst_mon.groupby("time.month")
sst_mon_anom = sst_mon_grouped - sst_mon_grouped.mean("time")
sst_mon_anom.attrs.update({"long_name": "SST anomaly", "units": "C"})
```

### 2. create NTI and SSTmD20

```python
nemo_nti = standardize(ssh_nemo_eq) + standardize(d20_eq)
```

```python
nemolin_nti = standardize(ssh_nemo_eq) + standardize(ssh_lin_eq)
```
