```python
import xarray as xr
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter,LatitudeFormatter
import matplotlib.dates as mdates
```

```python
ds = xr.open_zarr("data_fig2.zarr/")

sstmd20   = ds.sstmd20
nemo_nlti  = ds.nemo_nlti
nemolin_nlti = ds.nemolin_nlti
```

```python
width = 190*0.0393700787
fig = plt.figure(figsize=(width,5.5))
plt.rcParams['font.size'] =8
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})

ax1 = fig.add_subplot(131)
ax1.xaxis.set_major_formatter(LongitudeFormatter())
sstmd20.plot(
    y="time",
    vmin=-3, vmax=3,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
) 
year =  mdates.YearLocator(5,month=1) 
ax1.yaxis.set_major_locator(year)   
ax1.yaxis.set_major_formatter(mdates.DateFormatter('%Y')) # %Y-%m
plt.title("(a) SSTmD20", loc='left');
plt.title('')
plt.ylabel("time");
plt.xlabel('longitude');

ax2 = fig.add_subplot(132)
nemo_nlti.plot(
    y="time",
    vmin=-0.15, vmax=0.15,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
) 
ax2.xaxis.set_major_formatter(LongitudeFormatter())
ax2.yaxis.set_major_locator(year)   
ax2.yaxis.set_major_formatter(mdates.DateFormatter('%Y')) 
plt.title("(b) NEMO-NLTI",loc='left');
plt.title('')
plt.ylabel("");
plt.xlabel('longitude');

ax3 = fig.add_subplot(133)
nemolin_nlti.plot(
    y="time",
    vmin=-0.15, vmax=0.15,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
) 
ax3.xaxis.set_major_formatter(LongitudeFormatter())
ax3.yaxis.set_major_locator(year)   
ax3.yaxis.set_major_formatter(mdates.DateFormatter('%Y')) 
plt.title("(c) NEMOlin-NLTI",loc='left');
plt.title('')
plt.ylabel("");
plt.xlabel('longitude');

plt.subplots_adjust(wspace =0.5, hspace = 0)
plt.savefig("fig2.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
