```python
import xarray as xr
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter,LatitudeFormatter
```

```python
ds = xr.open_zarr('data_fig2.zarr/')
corr = ds.corr
mean_d20=ds.mean_d20
```

```python
from cartopy.mpl.ticker import LongitudeFormatter
y = corr.deptht
width = 95*0.0393700787
fig = plt.figure(figsize=(width,2.6))
plt.rcParams['font.size'] =8
from matplotlib import rc
# ax1 = fig.add_subplot(111)
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})
new_ticks =[0, 50,100,150,200,250,300] 
plt.yticks(new_ticks)

contourf = plt.contourf(corr.lon,corr.deptht,corr.T,[-1.0,-0.9,-0.8,-0.7,-0.6,-0.5,-0.4,-0.3,-0.2,-0.1,0,0.1],cmap=plt.cm.viridis_r) 
contour = plt.contour(corr.lon,corr.deptht,corr.T,[-0.8,-0.6,-0.4,-0.2,0],linewidths=1,colors='k')
mean_d20.plot( c="darkblue", ls="-", lw=3)
ax = plt.gca() 
ax.invert_yaxis()
ax.xaxis.set_major_formatter(LongitudeFormatter())
plt.title("",loc='left')
plt.title("")
plt.ylabel('Depth(m)')
plt.clabel(contour)
plt.colorbar(contourf)
plt.xlabel('longitude');

plt.savefig("fig2.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
