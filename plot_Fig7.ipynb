```python
import xarray as xr
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter,LatitudeFormatter
import matplotlib.dates as mdates
```

```python
ds = xr.open_zarr("data_fig7.zarr/")

nemo_nlti  = ds.nemo_nlti
nemolin_nlti = ds.nemolin_nlti
```

```python
width = 150*0.0393700787
fig = plt.figure(figsize=(width,5.5))
plt.rcParams['font.size'] =8
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})

ax1 = fig.add_subplot(121)
ax1.xaxis.set_major_formatter(LongitudeFormatter())
nemo_nlti.plot(
    y="time",
    vmin=-0.15, vmax=0.15,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
) 
year =  mdates.YearLocator(5,month=1) 
ax1.yaxis.set_major_locator(year)   
ax1.yaxis.set_major_formatter(mdates.DateFormatter('%Y')) # %Y-%m
plt.title("(a) NEMO", loc='left');
plt.title('')
plt.ylabel("time");
plt.xlabel('longitude');

ax2 = fig.add_subplot(122)
nemolin_nlti.plot(
    y="time",
    vmin=-0.15, vmax=0.15,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
) 
ax2.xaxis.set_major_formatter(LongitudeFormatter())
ax2.yaxis.set_major_locator(year)   
ax2.yaxis.set_major_formatter(mdates.DateFormatter('%Y')) 
plt.title("(b) Linear version 1",loc='left');
plt.title('')
plt.ylabel("");
plt.xlabel('longitude');


plt.subplots_adjust(wspace =0.5, hspace = 0)
plt.savefig("fig7.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
