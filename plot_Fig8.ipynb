```python
import xarray as xr
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter,LatitudeFormatter
import matplotlib.dates as mdates
```

```python
ds = xr.open_zarr('data_fig8.zarr')
```

```python
width = 190*0.0393700787
fig = plt.figure(figsize=(width,5.5))
plt.rcParams['font.size'] =8
from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})

ax1 = fig.add_subplot(131)
ax1.xaxis.set_major_formatter(LongitudeFormatter())
ds.nlti_nemo.plot(
    y="time", 
    vmin=-0.15, vmax=0.15,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
) 
year =  mdates.YearLocator(2,month=1) 
ax1.yaxis.set_major_locator(year)   
ax1.yaxis.set_major_formatter(mdates.DateFormatter('%Y')) # %Y-%m
plt.title("(a) NEMO",loc='left');
plt.title('');
plt.ylabel("time");
plt.xlabel('longitude');

ax2 = fig.add_subplot(132)
ax2.xaxis.set_major_formatter(LongitudeFormatter())
ds.nlti_nemolin.plot(
    y="time",
    vmin=-0.15, vmax=0.15,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
) 
ax2.yaxis.set_major_locator(year)   
ax2.yaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.title("(b) linear version 1",loc='left');
plt.title('');
plt.ylabel("");
plt.xlabel('longitude');

ax3 = fig.add_subplot(133)
ax3.xaxis.set_major_formatter(LongitudeFormatter())
ds.nlti_aviso.plot(
    y="time",
    vmin=-0.15, vmax=0.15,
    cmap=plt.cm.RdBu_r, extend="both",
    cbar_kwargs={'label': ''},
    rasterized=True
)
ax3.yaxis.set_major_locator(year)   
ax3.yaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.title("(c) linear version 2",loc='left');
plt.title('');
plt.ylabel("");
plt.xlabel('longitude');

ax3.annotate(text="",xy=(145,mdates.date2num(ds.time[231])),
             xytext = (212,mdates.date2num(ds.time[223])),
             xycoords='data',
             arrowprops=dict(facecolor='black', width=0.05,headwidth=3, headlength=4)
            )

ax3.annotate(text="",xy=(145,mdates.date2num(ds.time[215])),
             xytext = (212,mdates.date2num(ds.time[207])),
             xycoords='data',
             arrowprops=dict(facecolor='black', width=0.05,headwidth=3, headlength=4)
            )

ax3.annotate(text="",xy=(145,mdates.date2num(ds.time[184])),
             xytext = (212,mdates.date2num(ds.time[176])),
             xycoords='data',
             arrowprops=dict(facecolor='black', width=0.05,headwidth=3, headlength=4)
            )

ax3.annotate(text="",xy=(145,mdates.date2num(ds.time[76])),
             xytext = (212,mdates.date2num(ds.time[68])),
             xycoords='data',
             arrowprops=dict(facecolor='black',width=0.05,headwidth=3, headlength=4),
            )

ax3.annotate(text="",xy=(145,mdates.date2num(ds.time[88])),
             xytext = (212,mdates.date2num(ds.time[80])),
             xycoords='data',
             arrowprops=dict(facecolor='black',width=0.05,headwidth=3, headlength=4),
            )

ax3.annotate(text="",xy=(145,mdates.date2num(ds.time[127])),
             xytext = (212,mdates.date2num(ds.time[115])),
             xycoords='data',
             arrowprops=dict(facecolor='black',width=0.05,headwidth=3, headlength=4,),
            )

plt.tight_layout()
plt.savefig("fig8.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
