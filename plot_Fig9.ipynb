```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
from cartopy.mpl.ticker import LongitudeFormatter,LatitudeFormatter
import matplotlib.dates as mdates
```

```python
ds = xr.open_zarr('data_fig9.zarr/')
```

```python
width = 190*0.0393700787
```

```python
fig = plt.figure(figsize=(width, 5.5))
plt.rcParams['font.size'] =8

from matplotlib import rc
rc('font',**{'family':'sans-serif','sans-serif':['Helvetica']})

ax1 = fig.add_subplot(121)
ax1.xaxis.set_major_formatter(LongitudeFormatter())
ds.corr.plot(
    vmin=-1, vmax=1, cmap=plt.cm.RdBu_r,
    cbar_kwargs={'label': ''},
    rasterized=True
) 
ds.non_sig.plot.contourf(hatches=['.'], colors="None", add_colorbar=False) 
plt.title("(a) 21yr running cor(SST, D20)",loc='left')
plt.title('')
plt.xlim(140, 260)
plt.ylabel("time");
plt.xlabel('longitude');
plt.grid(color='grey',linestyle = ':')


ax2 = fig.add_subplot(122)
ax2.xaxis.set_major_formatter(LongitudeFormatter())
ds.slope.plot(
    vmin=0.0025, vmax=0.0045, cmap=plt.cm.YlOrBr,
    cbar_kwargs={'label': ''},
    rasterized=True
) 
plt.title("(b) Regression(D20,SSH)",loc='left') 
plt.title('')
plt.xlim(140, 260)
plt.ylabel("");
plt.xlabel('longitude');
plt.grid(color='grey',linestyle = ':')

plt.subplots_adjust(wspace =0.5, hspace = 0)
plt.savefig("fig9.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
